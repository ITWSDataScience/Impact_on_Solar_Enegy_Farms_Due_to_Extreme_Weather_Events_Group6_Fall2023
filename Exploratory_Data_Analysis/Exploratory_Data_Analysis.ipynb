```python
# Scatter plot of Risk Value vs Solar Radiance of Counties


plt.figure(figsize=(12, 6))
sns.scatterplot(x='RISK_VALUE', y='ANN', hue='COUNTY', data=df, palette='viridis')
plt.title('Scatter Plot of Risk Value vs Solar Radiance OF COUNTIES')
plt.xlabel('Risk Value')
plt.ylabel('Solar Radiance')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.savefig('scatter_plot.png')  # To saveSave the figure to a file
plt.show()


#This scatter plot clearly shows the counties with the highest Risk associated with Solar radiance
```

```python
#Bar Plot of Disaster Types by County
# Replace 'N' with NaN in the 'DISASTER' column
df['DISASTER'].replace('N', np.nan, inplace=True)

# Select specific disaster types to visualize
selected_disasters = ['Winter Storms', 'Earthquake', 'Storms', 'Drought', 'High winds', 'Wildfires', 'Rainstorms']

# Create a new column indicating whether each disaster type is present
for disaster in selected_disasters:
    df[disaster] = df['DISASTER'].str.contains(disaster, case=False, na=False)

# Melt the DataFrame to have a 'County' column, a 'Disaster Type' column, and a 'Count' column
melted_disaster_df = pd.melt(df, id_vars=['COUNTY'], value_vars=selected_disasters, var_name='Disaster_Type', value_name='Count')

# Grouped Bar Plot of Disaster Types by County
plt.figure(figsize=(14, 8))
sns.barplot(x='COUNTY', y='Count', hue='Disaster_Type', data=melted_disaster_df)
plt.title('Occurrences of Selected Disaster Types by County')
plt.xlabel('County')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Disaster Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#  Line Plot of Risk Values by Month
risk_columns = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
melted_df = pd.melt(df, id_vars=['COUNTY'], value_vars=risk_columns, var_name='Month', value_name='Risk_Value')

month_order = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
melted_df['Month'] = pd.Categorical(melted_df['Month'], categories=month_order, ordered=True)


plt.figure(figsize=(14, 8))
sns.lineplot(x='Month', y='Risk_Value', data=melted_df, hue='COUNTY', marker='o')
plt.title('Risk Values Over Months')
plt.xlabel('Month')
plt.ylabel('Risk Value')
plt.legend(title='County', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('line_plot_of_values_over_the_months.png')
plt.show()
```
